Importing Libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

working with the dataset

```python
# Generating a sequence of numbers from 1 to 100
sequence = np.array([i for i in range(1, 101)])

# Creating input-output pairs for training (use last 3 numbers to predict next one)
X, y = [], []
seq_length = 3  # using 3 numbers

for i in range(len(sequence) - seq_length):
    X.append(sequence[i:i+seq_length])
    y.append(sequence[i+seq_length])    # Output (next number)

# Converting to NumPy arrays
X, y = np.array(X), np.array(y)

# Reshaping X to fit LSTM input (samples, time steps, features)
X = X.reshape((X.shape[0], X.shape[1], 1))  # LSTM expects 3D input
```

building an LSTM model

```python
# Defining the LSTM model
model = Sequential([
    LSTM(50, activation='relu', input_shape=(3, 1)),  # 50 LSTM units
    Dense(25, activation='relu'),  # Extra Dense layer for better learning
    Dense(1)  # Single output neuron for next number
])

# Compiling the model
model.compile(optimizer='adam', loss='mse')  # Mean Squared Error for regression
```

Training the model

```python
model.fit(X, y, epochs=200, verbose=1)
```

Testing the model.
To predict the next number after [98, 99, 100] (should be 101).

```python
test_input = np.array([[98, 99, 100]]).reshape((1, 3, 1))

# Making prediction
prediction = model.predict(test_input)

# Rounding off the predicted value to the nearest integer
rounded_prediction = round(prediction[0][0])

# Print result
print("Predicted next number:", rounded_prediction)

```
